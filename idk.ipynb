```python
def custom_generator(data : object, index : int = 0):
    iterable = iter(data)
    for _ in range(len(data)):
        yield (index, iterable.__next__())
        index += 1
```

```python
for ahhhh in custom_generator(['a', 'b', 'c', 'd'], 3):
    print(ahhhh)
print()
for ahhhh in custom_generator({'a', 'b', 'c', 'd'}):
    print(ahhhh)
print()
for i in custom_generator({'a' : 1, 'b': 3}, 2):
    print(i)
```
